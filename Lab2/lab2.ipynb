```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

# Zadanie 1

```python
cywilny=pd.read_csv("data/stan_cywilny.csv",sep=";")
cywilny["stan_cywilny"]=cywilny["stan_cywilny"].astype("category")
cywilny["stan_cywilny"]=cywilny["stan_cywilny"].cat.codes
```

```python
cywilny
```

```python
x1 = np.array(cywilny["rok_studiow"])
y1 = np.array(cywilny["stan_cywilny"])
x1 = x1.reshape(-1,1)
model1 = LinearRegression()
model1.fit(x1, y1)
model1.predict(x1)
```

```python
model1.score(x1,y1)
```

```python
model2 = LogisticRegression()
model2.fit(x1, y1)
model2.predict_proba(x1)
```

```python
model2.predict(x1)
```

```python
model2.score(x1,y1)
```
